# Regional Housing Affordability Analysis

This notebook examines regional differences in housing affordability
across U.S. metropolitan areas using Zillow rental data.

The goal is to determine whether rent growth has occurred uniformly
across the country or whether certain regions have experienced
disproportionate increases.

## Load Zillow Rent Dataset

The Zillow Observed Rent Index (ZORI) dataset contains monthly rent
estimates for metropolitan areas across the United States.

For this analysis, we focus on identifying regional trends in rent
growth over time.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load Zillow rent dataset
rent_df = pd.read_csv(
    "Dataset #3 - Metro_zori_uc_sfrcondomfr_sm_month.csv"
)
```
